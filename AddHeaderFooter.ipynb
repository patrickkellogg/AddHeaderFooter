# AddHeaderFooter

Quick code to add a valid header and footer to text files and resave them as htm

```python
import os
import sys
import fnmatch

head1 = """<html>

<head>
<meta http-equiv="Content-Type" content="text/html; charset=windows-1252">
<meta name="GENERATOR" content="Microsoft FrontPage 12.0">
<meta name="ProgId" content="FrontPage.Editor.Document">
<title>"""
head2 = """</title>
<meta NAME="description" CONTENT=\""""
head3 = """\">
<meta NAME="keywords" CONTENT="Sex, porn, stories, gay, scat, leather, text">
</head>

<body>

<pre>"""
foot1 = """
</pre>
</body>
</html>
"""

mypath = "/Users/mudcub/Dropbox/English/"
# join filelist = [f for f in listdir(mypath) if isfile(join(mypath, f))]
#arr = os.listdir(mypath)
for dirpath, dirs, files in os.walk(mypath):
    for filelist in fnmatch.filter(files, '*.txt'):
        fin = open(os.path.join(mypath, filelist))
        if fin.mode == 'r':
            allcontents = fin.read()
            print(filelist)
            mytest = filelist
            #print(mytest)
            withoutname = mytest.rsplit('txt')[0]
            #print(withoutname)
            withoutdot = withoutname.rsplit('.')[0]
            #print(withoutdot)
            #print(contents)   
            together = head1 + withoutdot + head2 + withoutdot + head3 + allcontents + foot1
            #print(together)
            makenewfile = mypath + withoutdot + ".htm"
            fout = open(makenewfile, "w")
            fout.write(together)
            fout.write("")
            fout.close()
            fin.close()
            
```
